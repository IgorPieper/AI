```python
import pandas as pd
```

```python
car_sales = pd.read_csv("Data/car-sales-missing-data.csv")
print(car_sales)
```

```python
car_sales["Make"] = car_sales["Make"].str.lower()
print(car_sales)
```

```python
## Filling blank spots
car_sales["Odometer"] = car_sales["Odometer"].fillna(0)
print(car_sales)
```

```python
## Deliting rows, where is NAN
car_sales = car_sales.dropna()
print(car_sales)
```

```python
## Adding column
car_sales["Seats"] = [5, 5, 5, 5, 5, 5]
print(car_sales)
```

```python
# Deleting column
del car_sales["Seats"]
print(car_sales)
```

```python
# Random samples from data (50% of all)
shuffled_car_sales = car_sales.sample(frac=0.5)
print(shuffled_car_sales)
```

```python
# Sorting indexes
shuffled_car_sales = shuffled_car_sales.reset_index(drop=True)
print(shuffled_car_sales)
```

```python
# Changes Kilometers to Miles
car_sales["Odometer"] = car_sales["Odometer"].apply(lambda x:x / 1.6)
print(car_sales)
```
