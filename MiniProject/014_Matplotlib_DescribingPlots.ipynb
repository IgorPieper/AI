```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1, 2, 3, 4, 5]
y = [50, 100, 200, 250, 400]
```

```python
fig, ax = plt.subplots()
ax.plot(x, y);
```

```python
# Chart with descriptions
fig, ax = plt.subplots()
ax.set(title = "title",
      xlabel = "x label",
      ylabel = "y label")
ax.plot(x, y);
```

```python
# Save Chart
fig.savefig("Images/chart.png")
```
