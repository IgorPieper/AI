```python
# Imports
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Classificator
heart_disease = pd.read_csv("Data/heart-disease.csv")
heart_disease.head()

x1 = heart_disease.drop("target", axis=1)
y1 = heart_disease["target"]

from sklearn.model_selection import train_test_split
x_train1, x_test1, y_train1, y_test1 = train_test_split(x1, y1, test_size=0.2)

from sklearn.ensemble import RandomForestClassifier
clf1 = RandomForestClassifier()
clf1.fit(x_train1, y_train1)
y_pred1 = clf1.predict(x_test1)
```

```python
# Confusion Matrix  (Don't work on Regressor)
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test1, y_pred1)
```

```python
pd.crosstab(y_test1, y_pred1, rownames=["Actual Labels"], colnames=["Predicted Labels"])
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(estimator=clf1, X=x1, y=y1)
```

```python
ConfusionMatrixDisplay.from_predictions(y_true=y_test1, y_pred=y_pred1)
```
