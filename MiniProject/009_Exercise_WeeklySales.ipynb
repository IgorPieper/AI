```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(0)

sindex = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun",]
scolumns = ["Almond Butter", "Peanat Butter", "Cashhew Butter"]

sales_amounts = np.random.randint(20, size=(7, 3))
sales_amounts
```

```python
weekly_sales = pd.DataFrame(sales_amounts, index=sindex, columns=scolumns)
weekly_sales
```

```python
prices = np.array([15, 10, 20])
prices
```

```python
butter_prices = pd.DataFrame(prices.reshape(1, 3), index=["Price"], columns=scolumns)
butter_prices
```

```python
total_sales = prices.dot(sales_amounts.T)
total_sales
```

```python
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```
