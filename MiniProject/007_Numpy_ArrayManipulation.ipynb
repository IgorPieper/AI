```python
import numpy as np
```

```python
a1 = np.array([1, 2, 3])
a1
```

```python
a2 = np.array([[1, 2, 3], [4, 5, 6]])
a2
```

```python
a3 = np.array([[[0.5, 1.5, 2.5]], [[0.25, 0.5, 0.75]]])
a3
```

```python
# Selecting a data
a3[0][0][0]
```

```python
# Selecting a column
a3[:,:,2:]
```

```python
# Adding arrays
a4 = a1 + [3, 2, 1]
a4
```

```python
# Subtraction
a4 = a4 - a1
a4
```

```python
# Multiplication

a5 = a2 * a4
a5
```

```python
# Floor
print(a5 // 2)
```

```python
# Modulo
print(a5 % 2)
```

```python
# Mean - The average value of a set of numbers.
print(np.mean(a5))
```

```python
# Variance - How spread out the numbers are from the mean.
print(np.var(a5))
```

```python
# Standard Deviation - The typical distance each number is from the mean.
print(np.std(a5))
```
