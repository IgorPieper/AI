```python
# Imports
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
car_sales = pd.read_csv("Data/car-sales-extended-missing-data.csv")
car_sales.head()
```

```python
# We have missing data!
car_sales.isna().sum()
```

```python
# We can implement data
car_sales["Make"] = car_sales["Make"].fillna("missing")
car_sales["Colour"] = car_sales["Colour"].fillna("missing")
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].fillna(0)
car_sales["Doors"] = car_sales["Doors"].fillna(0)
```

```python
# We can delete rows with missing data
car_sales.dropna(inplace=True)
```
