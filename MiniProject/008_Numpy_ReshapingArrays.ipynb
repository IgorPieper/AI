```python
import numpy as np
```

```python
a2 = np.array([[1, 2, 3], [4, 5, 6]])
a2
```

```python
a3 = np.array([[[0.5, 1.5, 2.5]], [[0.25, 0.5, 0.75]]])
a3
```

```python
a4 = a2.reshape(2, 3, 1)
a4
```

```python
a5 = a4 * a3
a5
```

```python
# Transpose
a5.T
```

```python
# Matrix Multiplication
np.dot(a4, a5)
```
